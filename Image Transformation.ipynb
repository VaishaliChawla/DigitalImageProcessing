### Adding value to each pixel

```python
file = 'picture.jpeg'
im = Image.open(file)
imp = p.imshow(im)
p.axis("off")
p.show()
for y in range(im.height):
    for x in range(im.width):
        r,g,b = im.getpixel((x,y))
        R = r+50
        G = g+4
        B = b+2
        new = (R,G,B)
        im.putpixel((x,y), new)
im.save('picnew.jpg')

p.imshow(im)
p.axis("off")
p.show()
```

### Log transformation

```python

# Load the image
img = cv2.imread('gray.jpeg')
p.imshow(img)
p.axis("off")
p.show()
# Apply log transform
img_log = (np.log(img+1)/(np.log(1+np.max(img))))*255

# Specify the data type
img_log = np.array(img_log,dtype=np.uint8)

# Display the image
#cv2.imshow('log_image',img_log )
#cv2.imshow('original_img',img)
#cv2.waitKey(0)

p.imshow(img_log)
p.axis("off")
p.show()
```

### Gamma correction

```python
#Gamma Correction on an RGB image

im = cv2.imread("bird.jpg", cv2.COLOR_BGR2RGB) #openCV uses BGR scheme, so we convert it to RGB to avoid image color distortion
im2 = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
p.imshow(im2)
p.axis("off")
p.show()
g1 = 2.1
im2_dash = np.array(255 * (im2/255) ** g1, dtype=np.uint8)
p.imshow(im2_dash)
p.axis("off")
p.show()
```
